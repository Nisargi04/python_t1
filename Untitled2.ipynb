```python
import pandas as pd
def find_outlier(ds,col):
    quart1=ds[col].quantile(0.25)
    quart3=ds[col].quantile(0.75)
    iqr=quart3-quart1
    low_val=quart1-1.5*iqr
    print(low_val)
    high_val=quart3+1.5*iqr
    print(high_val)
    ds=ds.loc[(ds[col]<low_val)| (ds[col]>high_val)]
    return ds
df=pd.read_csv('auto-mpg.csv')
find_outlier(df,'mpg')
```

```python
find_outlier(df,'acceleration')
```

```python
import pandas as pd
def remove(ds,col):
    quart1=ds[col].quantile(0.25)
    quart3=ds[col].quantile(0.75)
    iqr=quart3-quart1
    low_val=quart1-1.5*iqr
    print(low_val)
    high_val=quart3+1.5*iqr
    print(high_val)
    ds=ds.loc[(ds[col]>=low_val) & (ds[col]<=high_val)]
    return ds
```

```python
find_outlier(df,'mpg')
remove(df,'mpg')
```

```python
import pandas as pd
#function that accepts standar arguments
def add_values(x,y,z):
    return x+y+z
df=pd.DataFrame({'a':[1,2],'b':[10,20]})
#use the 'args' paraneter(as tuple) in apply()
#a[1] sathe args ni value parameters ma run thashe same agal pn
df1=df.apply(add_values,args=(1,2))
print(df1)
```

```python
import pandas as pd
#function that accepts standar arguments
def multiply_by_m(x,m):
    return x*m
df=pd.DataFrame({'a':[1,2],'b':[10,20]})
#use the 'args' paraneter(as tuple) in apply()
#'x is implicity passed by apply()(each row/col value) and 'y' and 'z' are args
df1=df.apply(multiply_by_m,m=10)
print(df1)
```

```python
import pandas as pd
df1=pd.DataFrame({"employee":['a',"b","c"],
                 'department':['it','hr','finance']})
df2=pd.DataFrame({"employee":['a',"b",'d'],
                 'salary':[50000,40000,30000]})
result=pd.merge(df1,df2,on='employee',how='inner')
print(result)
```

```python
result1=pd.merge(df1,df2,on='employee',how='outer')
print(result1)
```

```python
result1=pd.merge(df1,df2,on='employee',how='left')#first dataframe no data ne ema merge
print(result1)
```

```python
result1=pd.merge(df1,df2,on='employee',how='right')#second dataframe pramne merging
print(result1)
```

```python
df=pd.DataFrame({"region":["east","east","west","west","east","west"],
                "sales":[100,150,200,120,110,130],
                "manager":["a","b","c","d","e","f"]})
df
```

```python
def sales_range(group):
    return group["sales"].max()-group['sales'].min()
sales=df.groupby('region').apply(sales_range)
print(sales)
```

```python
import pandas as pd
df=pd.DataFrame({"dept":["sales","sales","hr","hr","it","it"],
                 "salary":[50000,60000,40000,45000,70000,75000],
                 "bonus":[5000,6000,2000,2500,8000,8500]})
df
```

```python
result=df.groupby('dept').agg('mean')
print(result)
```

```python
result=df.groupby('dept').agg(['mean','min','max'])
print(result)
```

```python
result=df.groupby('dept').agg({"salary":['mean','max'],"bonus":"sum"})
print(result)
```

```python
result=df.groupby('dept').agg(meansalary=("salary","mean"),maxsalary=("salary","max"),totalbonus=("bonus","sum"))
print(result)
```
