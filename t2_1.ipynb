```python
import pandas as pd
df=pd.read_csv("data.csv")
```

```python
df.head()

```

```python
x=df.iloc[:,0:1]#x must be dataframe
y=df.iloc[:,-1]#y must be series

type(y)
```

```python
type(x)
```

```python
print(x.shape)
print(y.shape)
```

```python
#split the data in training testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#creating linearRegression model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
y_pred=lr.predict(x)
print(y_pred)
```

```python
diff=pd.DataFrame({"Actual":y,"Predicted":y_pred})
diff
```

```python
print("Coefficient",lr.coef_)#m :- y=mx+c multiple values avi shake as m so that's why it is represented as array
print("intercept:",lr.intercept_)#c
```

```python
from sklearn import metrics
print("MAE: ",metrics.mean_absolute_error(y,y_pred))
print("MSE: ",metrics.mean_squared_error(y,y_pred))
print("R2 score: ",metrics.r2_score(y,y_pred))
```

```python
y_pred=lr.predict([[30]])#[30] karshu to error cause x must be 2D
print(y_pred)
```
